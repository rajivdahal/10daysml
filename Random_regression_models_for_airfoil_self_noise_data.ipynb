**Applying Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
regressor = LinearRegression()
```

```python
mse = cross_val_score(regressor,X,y,scoring='neg_mean_squared_error',cv=5)
```

```python
mse
```

Above MSE suggests that Modal is somewhat unstable with the data we have for linear regression

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

**Testing with Ridge regressor**

```python
from sklearn.linear_model import Ridge
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
```

```python
parameters = {'alpha':[1e-15,1e-8,1e-3,1e-2,1,5,10,20,30,40,50]}
#To select best lambda value we use GridSearchCV
ridge_regressor = GridSearchCV(ridge,parameters, scoring="neg_mean_squared_error", cv=5, verbose=1)
```

```python
ridge_regressor.fit(X,y)
```

```python
ridge_regressor.best_params_
```

```python
ridge_regressor.best_score_
```

```python
y_pred = ridge_regressor.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

**Using Decision tree**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
decision_tree_regressor = DecisionTreeRegressor()
```

```python
parameters = {'max_depth':[10,20,50,60,70,100,1300], 'min_samples_split':[2,3,4,6,10,12],'min_samples_leaf':[1,2,3,4,5,6,7,8]}
#To select best lambda value we use GridSearchCV
decision_tree_regressorModal = GridSearchCV(decision_tree_regressor,parameters, cv=5, verbose=1)
```

```python
decision_tree_regressorModal.fit(X_train,y_train)
```

```python
y_pred= decision_tree_regressorModal.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```

```python
# Finally decision tree has got good score of 86%
```

**Using Random forest**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_forest_regressor = RandomForestRegressor(criterion='squared_error')
```

```python
parameters = {'n_estimators':[100,200,150,120],'max_depth': [None, 5, 10, 20, 30, 50], 'min_samples_split':[1,2,3,4,5], 'min_samples_leaf':[1,2,3,4]}
```

```python
random_forest_regressorModal = GridSearchCV(random_forest_regressor,parameters, cv=5, verbose=1)
```

```python
random_forest_regressorModal.fit(X_train,y_train)
```

```python
y_pred=random_forest_regressorModal.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

**The accuracy for Random Forest is 93% so we concluded that the best model is Random forest model among all. But we can still increase the accuracy of model by hyperparameter tuning**
