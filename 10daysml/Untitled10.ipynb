```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("data.csv")
df
```

```python
df.isnull().sum()
```

```python

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.loc[:,["Brand","Model_year","Transmission","Fuel_type"]],df.loc[:,["Price"]]
                                                ,test_size=0.2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(sparse=False,drop='first'),['Brand','Transmission','Fuel_type'])
],remainder='passthrough')
```

```python
transformer.fit(X_train)
```

```python
X_train_new=transformer.transform(X_train)
X_test_new=transformer.transform(X_test)
```
