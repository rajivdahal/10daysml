```python
from sklearn.model_selection import train_test_split
```

# selecting X and Y column and mentioning the test size

```python
X_train,X_test,y_train,y_test=train_test_split(newnewdf.iloc[:,0:2],newnewdf.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG',"PG"]])
```

```python
oe.fit(X_train)
```

```python
X_train1=oe.transform(X_train)
X_test1=oe.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train1,columns=X_train.columns)
```
