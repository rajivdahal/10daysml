```python
#creating my own stopwords list
from nltk.corpus import stopwords
imported_stopwords = stopwords.words('english')
for word in ['not','nor']:
  if word in imported_stopwords:
      imported_stopwords.remove(word)
print(imported_stopwords)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import sent_tokenize,word_tokenize
nltk.download('punkt')
nltk.download('wordnet')
corpus = []
lemmetizedCorpus = []
paragraph = "I love learning data science. I am just testing this feature"
stem_tokenized_paragraph = sent_tokenize(paragraph)
imported_stopwords = stopwords.words('english')
for i in range(len(stem_tokenized_paragraph)):
  tokenizedWords = word_tokenize(stem_tokenized_paragraph[i])
  stemmedWords = [PorterStemmer().stem(word) for word in tokenizedWords if not word in imported_stopwords]
  print('stemword is', stemmedWords)
  corpus.append(' '.join(stemmedWords))
print('corpus is', corpus)

#the same thing you can do using lemmetizer

for i in range(len(stem_tokenized_paragraph)):
  tokenizedWords = word_tokenize(stem_tokenized_paragraph[i])
  lemmetizedWords = [WordNetLemmatizer().lemmatize(word) for word in tokenizedWords if not word in imported_stopwords]
  print('stemword for lemmetized is', lemmetizedWords)
  lemmetizedCorpus.append(' '.join(lemmetizedWords))
print('lemmetizedCorpus is', lemmetizedCorpus)
```

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import sys
def vectorizeParagraph(paragraph, vectorizer):
  nltk.download('punkt')
  nltk.download('wordnet')
  nltk.download('stopwords')
  paragraph_sentence_list = sent_tokenize(paragraph)
  corpus = []
  imported_stopwords = stopwords.words('english')
  for i in range(len(paragraph_sentence_list)):
    cleaned_sentence = re.sub('[^a-zA-Z0-9]',' ', paragraph_sentence_list[i])
    tokenized_words = word_tokenize(cleaned_sentence)
    stemmed_words = [WordNetLemmatizer().lemmatize(word) for word in tokenized_words if not word in imported_stopwords]
    corpus.append(' '.join(stemmed_words))

  #we are going to apply bag of words in corpus to this we need to use countvectorizer from sklearn feature extraction
  print('corpus is', corpus)

  vectorized = vectorizer.fit_transform(corpus)
  np.set_printoptions(threshold=sys.maxsize)
  print(vectorized.toarray())
  return vectorized


vectorizer = CountVectorizer()
vectorizeParagraph(paragraph, vectorizer)


#problem here is it is not that robust in


```

```python
def getBagOfWordsFeaturesNames(vectorizer):
  return vectorizer.get_feature_names_out()
def getVocabulary(vectorizer):
  return vectorizer.vocabulary_
```

```python
vectorizer = CountVectorizer(ngram_range= (1, 1))
paragraph = 'good boy. good girl. Boy and girl good'
vectorizeParagraph(paragraph, vectorizer)
```
