```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
import re
```

```python
corpus = []
for i in range(0,len(dataset)):
  review = re.sub(r'<br\s*/?>|[^a-zA-Z0-9]', ' ',dataset['review'][i])
  review = review.lower()
  review = review.split()
  review=[ps.stem(word) for word in review if not word in stopword_list]
  review = ' '.join(review)
  corpus.append(review)
  # re.sub('[a-zA-Z0-9]',' ', dataset['review'][i])
```

```python
corpus
```

```python
##Train test split
```

```python
##independent features and dependent features
y = pd.get_dummies(dataset['sentiment'], drop_first=True)
y=y.astype(int)
# Get dummies gets the dummy categorical variables in boolean- True or False
#here in Y value there are 2 datasets - Positive and Negative, get_dummies maps Positive and Negative values on 2 columns stating either True or False
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(corpus, y, test_size=0.20,random_state=0)
```

```python
###Create bag of words
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,ngram_range=(1,2))
X_train=cv.fit_transform(X_train).toarray()
```

```python
X_test=cv.transform(X_test).toarray()
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier().fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```
