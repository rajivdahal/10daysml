**Pre pruning with grid search cv and cross valscore**

Basically prepruning is finding best hyperparameters for a decision tree to be more accurate

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size=0.20, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
#max_depth = prunes the tree
#min_samples_split = controls splitting the split if it is on the given threashold eg. If min_samples_split = 4 then, it will split to another node if the node has min 5 samples, else the node becomes leaf node
#In short: minimum number of samples required to split the node
#min_samples_leaf = minimum number of samples required to become a leaf node
parameter = {'max_depth':[2.4,6,8,10], 'min_samples_split':[2,3,4], 'min_samples_leaf':[1,2,3]}
classifier = DecisionTreeClassifier()
decisionTreeClassifier = GridSearchCV(classifier,parameter,cv=5)
decisionTreeClassifier.fit(X_train,y_train)
```

```python
#calculating the accuracy
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_curve, f1_score
```

```python
y_pred = decisionTreeClassifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```
